```python
# preamble.py 파일 내 코드
from IPython.display import set_matplotlib_formats, display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mglearn
from cycler import cycler

set_matplotlib_formats('pdf', 'png')
plt.rcParams['figure.dpi'] = 300
plt.rcParams['image.cmap'] = "viridis"
plt.rcParams['image.interpolation'] = "none"
plt.rcParams['savefig.bbox'] = "tight"
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['legend.numpoints'] = 1
plt.rc('axes', prop_cycle=(
    cycler('color', mglearn.plot_helpers.cm_cycle.colors) +
    cycler('linestyle', ['-', '-', "--", (0, (3, 3)), (0, (1.5, 1.5))])))

np.set_printoptions(precision=3, suppress=True)

pd.set_option("display.max_columns", 8)
pd.set_option('precision', 2)

__all__ = ['np', 'mglearn', 'display', 'plt', 'pd']
```

```python
# 데이터셋을 만듭니다
# forge(구축하다)
X, y = mglearn.datasets.make_forge()

# 산점도를 그립니다
# 이 산점도의 점 하나가 각 데이터 포인트(샘플=개체)를 나타냅니다
# x축은 첫 번째 특성(feature), y축은 두 번째 특성(feature)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["클래스 0", "클래스 1"], loc=4)
plt.xlabel("1 feature")
plt.ylabel("2 feature")

# forge 데이터셋의 산점도
# X.shape에서 이 데이터셋은 데이터 포인트 26개와 특성 2개를 가진다는 것을 알 수 있음
print("X.shape:", X.shape)
```

```python
# 회귀 알고리즘 설명을 위해 인위적으로 만든 wave 데이터셋을 사용
X, y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("feature")
plt.ylabel("target")
```

```python
# scikit-learn에 들어 있는 실제 데이터셋도 두 개를 사용
# Wisconsin Breast Cancer 데이터셋
# 레이블은 malignant, benign
# 종양이 악성인지? 를 예측할 수 있도록 학습하는 것이 과제

from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print("cancer.keys():\n", cancer.keys())

# 이 데이터셋은 
# 569개 데이터 포인트 
# 30개 특성 
print("유방암 데이터의 형태:", cancer.data.shape)
# 212개 악성, 357개 양성
print("클래스별 샘플 갯수:\n",
      {n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))})
print("특성 이름:\n", cancer.feature_names)
```

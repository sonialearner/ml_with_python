```python
import matplotlib
import matplotlib.pylab as pltlab

# 한글 깨짐 해결
pltlab.rcParams["font.family"] = "Malgun Gothic"
pltlab.rcParams["font.size"] = 10
# plt.rcParams["font.figsize"] = (14,4)

# 마이너스 깨짐 해결
matplotlib.rcParams['axes.unicode_minus'] = False

# k-최근접 이웃 알고리즘은 회귀(regression) 분석에도 쓰임
# wave 데이터셋 이용하여
# 이웃이 하나인 K-NN 사용

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# x축: 초록색 3개의 별들: 테스트 데이터 3개
mglearn.plots.plot_knn_regression(n_neighbors=1)

# 최근접 이웃을 한 개만 이용할 때
# 예측(prediction)은? 그냥 가장 가까운 이웃의 target값
# 예측(prediction)은 파랑색 3개의 별들
```

```python
# 이웃을 2개 이상 사용하여
# 회귀(regression) 분석을 할 수 있음
# 여러 개의 최근접 이웃을 사용할 땐 이웃 간의 평균이 예측됨

mglearn.plots.plot_knn_regression(n_neighbors=3)
```

```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

# wave 데이터셋을 
# 훈련 세트와 테스트 세트로 나눔
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# 이웃의 수를 3으로 하여
# 모델의 객체를 만듦
reg = KNeighborsRegressor(n_neighbors=3)

# 훈련 데이터와 타깃을 사용하여 
# 모델을 학습시킴 (fit 메서드 사용)
reg.fit(X_train, y_train)

# weights 매개변수가 기본값 uniform 일 때는
# np.mean 함수를 사용하여, 단순 평균을 계산
# distance일 때는 거리를 고려한 가중치 평균을 계산
```

```python
# predict 메서드 사용
print("테스트 세트 예측:\n", reg.predict(X_test))

# score 메서드 사용하여 모델을 평가함
# 이 메서드는 회귀(regression)일 땐, R의 제곱한 값을 반환함
print("테스트 세트 R^2: {:.2f}".format(reg.score(X_test, y_test)))

# R의 제곱 값은 '결정 계수'라고도 함
# 이 값은 회귀 모델에서 예측 적합도를 0과 1 사이의 값으로 계산한 것
# 1은? 예측이 완벽한 경우
# 0은 훈련 세트의 출력값(y_train)의 평균으로만 예측하는 모델의 경우

# R^2값이 0.83이므로 모델이 비교적 잘 들어맞은 것
```

```python
# KNeighborsRegressor 분석
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# -3 과 3 사이에 1,000 개의 데이터 포인트를 만듭니다
line = np.linspace(-3, 3, 1000).reshape(-1, 1)

for n_neighbors, ax in zip([1, 3, 9], axes):
    # 1, 3, 9 이웃을 사용한 예측을 합니다
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)

    ax.set_title(
        "{} 이웃의 훈련 스코어: {:.2f} 테스트 스코어: {:.2f}".format(
            n_neighbors, reg.score(X_train, y_train), reg.score(X_test, y_test)))
    ax.set_xlabel("특성")
    ax.set_ylabel("타깃")
axes[0].legend(["모델 예측", "훈련 데이터/타깃", "테스트 데이터/타깃"], loc="best")
```

```python
# 이웃을 많이 사용하면
# 훈련 데이터에는 잘 안 맞을 수 있지만
# '더 안정된 예측'을 얻게 됨

# 이웃을 1개만 사용할 때
# 훈련 세트의 각 데이터 포인트가
# 예측에 주는 영향이 커서
# 예측값이 훈련 데이터 포인트를 모두 지나감
```
